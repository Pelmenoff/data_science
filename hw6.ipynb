```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data_2d = pd.read_csv('data_2d.csv')

scaler = StandardScaler()
data_2d_scaled = scaler.fit_transform(data_2d)

inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_2d_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.title('Ліктьовий метод для визначення оптимальної кількості кластерів')
plt.show()
```

```python
kmeans_2d = KMeans(n_clusters=3, random_state=25)
clusters_2d = kmeans_2d.fit_predict(data_2d_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(data_2d_scaled[:, 0], data_2d_scaled[:, 1], c=clusters_2d, cmap='viridis')
plt.scatter(kmeans_2d.cluster_centers_[:, 0], kmeans_2d.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Перша ознака')
plt.ylabel('Друга ознака')
plt.title('Кластеризація двовимірного датасету з 3 кластерами')
plt.legend()
plt.show()
```

```python
mnist = pd.read_csv('mnist.csv')

mnist_scaled = scaler.fit_transform(mnist)

inertia_mnist = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(mnist_scaled)
    inertia_mnist.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(K, inertia_mnist, 'bx-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.title('Ліктьовий метод для визначення оптимальної кількості кластерів для MNIST')
plt.show()
```

```python
kmeans_mnist = KMeans(n_clusters=8, random_state=25)
clusters_mnist = kmeans_mnist.fit_predict(mnist_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(mnist_scaled[:, 0], mnist_scaled[:, 1], c=clusters_mnist, cmap='viridis', s=1)
plt.scatter(kmeans_mnist.cluster_centers_[:, 0], kmeans_mnist.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Перша ознака')
plt.ylabel('Друга ознака')
plt.title('Кластеризація MNIST з 8 кластерами')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
mnist_pca = pca.fit_transform(mnist_scaled)

kmeans_mnist_pca = KMeans(n_clusters=8, random_state=25)
clusters_mnist_pca = kmeans_mnist_pca.fit_predict(mnist_pca)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(mnist_pca[:, 0], mnist_pca[:, 1], c=clusters_mnist_pca, cmap='viridis', s=1)
plt.scatter(kmeans_mnist_pca.cluster_centers_[:, 0], kmeans_mnist_pca.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Перша головна компонента')
plt.ylabel('Друга головна компонента')
plt.title('Кластеризація MNIST з 8 кластерами після PCA')
plt.legend()
plt.show()
```
